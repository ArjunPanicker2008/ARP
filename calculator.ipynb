```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Display and Output
display_box = widgets.Text(value='', placeholder='0', disabled=False, layout=widgets.Layout(width='100%', height='40px'))
output = widgets.Output()

# Button labels
buttons = [
    ['7', '8', '9', '/'],
    ['4', '5', '6', '*'],
    ['1', '2', '3', '-'],
    ['0', '.', 'C', '+'],
    ['(', ')', '=', '**']
]

# Function to handle button clicks
def on_button_click(b):
    with output:
        if b.description == 'C':
            display_box.value = ''
        elif b.description == '=':
            try:
                result = eval(display_box.value)
                display_box.value = str(result)
            except Exception as e:
                display_box.value = 'Error'
        else:
            display_box.value += b.description

# Create and display buttons
button_widgets = []
for row in buttons:
    button_row = []
    for label in row:
        btn = widgets.Button(description=label, layout=widgets.Layout(width='60px', height='40px'))
        btn.on_click(on_button_click)
        button_row.append(btn)
    button_widgets.append(widgets.HBox(button_row))

# Show UI
display(display_box)
for row in button_widgets:
    display(row)
display(output)
```
