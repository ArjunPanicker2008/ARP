```python
import os
import shutil

def list_files():
    print("\nFiles in current directory:")
    for item in os.listdir():
        print(f" - {item}")

def create_folder():
    name = input("Enter folder name: ")
    try:
        os.makedirs(name)
        print(f"Folder '{name}' created.")
    except FileExistsError:
        print("Folder already exists.")

def move_file():
    src = input("Enter file name to move: ")
    dst = input("Enter destination folder: ")
    try:
        shutil.move(src, dst)
        print(f"Moved '{src}' to '{dst}'.")
    except Exception as e:
        print("Error:", e)

def delete_file():
    name = input("Enter file name to delete: ")
    try:
        os.remove(name)
        print(f"Deleted '{name}'.")
    except Exception as e:
        print("Error:", e)

def main():
    while True:
        print("\nFile Manager Options:")
        print("1 - List files")
        print("2 - Create folder")
        print("3 - Move file")
        print("4 - Delete file")
        print("5 - Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            list_files()
        elif choice == "2":
            create_folder()
        elif choice == "3":
            move_file()
        elif choice == "4":
            delete_file()
        elif choice == "5":
            print("Goodbye.")
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main()
```
