```python
# --- Stock Manager (Single Cell, GitHub-safe) ---
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
import json

# Prevent broken widget metadata on export
try:
    from google.colab import _message
    _message.blocking_request('remove_widget_metadata', {})
except:
    pass  # Safe outside Colab or already clean

# In-memory stock store
stock = []

# Widgets
item = widgets.Text(description="Item:")
qty = widgets.IntText(description="Qty:")
price = widgets.FloatText(description="Price:")
out = widgets.Output()

# Buttons
add_btn = widgets.Button(description="Add/Update", button_style='success')
view_btn = widgets.Button(description="View", button_style='info')
clear_btn = widgets.Button(description="Clear", button_style='danger')

# Functions
def add_update(b):
    name = item.value.strip()
    q = qty.value
    p = price.value
    if not name:
        return
    for s in stock:
        if s['Item'].lower() == name.lower():
            s['Quantity'] += q
            s['Price'] = p
            break
    else:
        stock.append({'Item': name, 'Quantity': q, 'Price': p})
    item.value, qty.value, price.value = '', 0, 0.0
    show()

def show(b=None):
    with out:
        clear_output()
        if stock:
            display(pd.DataFrame(stock))
        else:
            print("Stock is empty.")

def clear_all(b):
    stock.clear()
    with out:
        clear_output()
        print("Stock cleared.")

# Events
add_btn.on_click(add_update)
view_btn.on_click(show)
clear_btn.on_click(clear_all)

# UI layout
display(widgets.HTML("<h3>ðŸ“¦ Simple Stock Manager</h3>"))
display(item, qty, price, widgets.HBox([add_btn, view_btn, clear_btn]), out)
```
