```python
# --- 📦 Stock Manager
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# ⚠️ Try removing widget metadata (Colab only; safe to ignore elsewhere)
try:
    from google.colab import _message
    _message.blocking_request('remove_widget_metadata', {})
except:
    pass

# ---- Data Store ----
stock = []

# ---- Widgets ----
item = widgets.Text(description="Item:", placeholder="e.g. Pen")
qty = widgets.BoundedIntText(description="Qty:", min=0, max=10000, value=0)
price = widgets.BoundedFloatText(description="Price:", min=0.0, max=1e6, value=0.0)
out = widgets.Output()

btn_add = widgets.Button(description="Add/Update", button_style="success")
btn_view = widgets.Button(description="View Stock", button_style="info")
btn_clear = widgets.Button(description="Clear All", button_style="danger")

# ---- Functions ----
def add_update_stock(b):
    name = item.value.strip()
    q = qty.value
    p = price.value
    if not name:
        return
    for s in stock:
        if s['Item'].lower() == name.lower():
            s['Quantity'] += q
            s['Price'] = p
            break
    else:
        stock.append({'Item': name, 'Quantity': q, 'Price': p})
    item.value, qty.value, price.value = '', 0, 0.0
    view_stock()

def view_stock(b=None):
    with out:
        clear_output()
        if stock:
            df = pd.DataFrame(stock)
            df['Total'] = df['Quantity'] * df['Price']
            display(df)
            print(f"📦 Total Inventory Value: ${df['Total'].sum():,.2f}")
        else:
            print("No items in stock.")

def clear_stock(b):
    stock.clear()
    with out:
        clear_output()
        print("✅ Stock cleared.")

# ---- Button Actions ----
btn_add.on_click(add_update_stock)
btn_view.on_click(view_stock)
btn_clear.on_click(clear_stock)

# ---- Display UI ----
display(widgets.HTML("<h3>🧾 Simple Stock Management</h3>"))
display(item, qty, price)
display(widgets.HBox([btn_add, btn_view, btn_clear]))
display(out)
```
