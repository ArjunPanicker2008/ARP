```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

# Data store
stock_data = []

# Input widgets
item_input = widgets.Text(placeholder="Item name", description="Item:")
qty_input = widgets.IntText(value=0, description="Qty:")
price_input = widgets.FloatText(value=0.0, description="Price:")
output = widgets.Output()

# Action buttons
add_btn = widgets.Button(description="Add/Update", button_style='success')
view_btn = widgets.Button(description="View Stock", button_style='info')
clear_btn = widgets.Button(description="Clear All", button_style='danger')

# Functions
def add_item(b):
    name = item_input.value.strip()
    qty = qty_input.value
    price = price_input.value
    if not name:
        return
    for item in stock_data:
        if item['Item'].lower() == name.lower():
            item['Quantity'] += qty
            item['Price'] = price
            break
    else:
        stock_data.append({'Item': name, 'Quantity': qty, 'Price': price})
    item_input.value = ''
    qty_input.value = 0
    price_input.value = 0.0
    show_stock()

def show_stock(b=None):
    with output:
        clear_output()
        if stock_data:
            df = pd.DataFrame(stock_data)
            display(df)
        else:
            print("No stock data available.")

def clear_stock(b):
    stock_data.clear()
    with output:
        clear_output()
        print("Stock cleared.")

# Bind buttons
add_btn.on_click(add_item)
view_btn.on_click(show_stock)
clear_btn.on_click(clear_stock)

# Layout
form = widgets.VBox([
    widgets.HTML("<h3>ðŸ§¾ Stock Management System</h3>"),
    item_input, qty_input, price_input,
    widgets.HBox([add_btn, view_btn, clear_btn]),
    output
])
display(form)
```
