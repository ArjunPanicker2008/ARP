```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

# Data storage
stock_data = []

# Widgets
item_name = widgets.Text(placeholder="Item name", description="Item:")
quantity = widgets.IntText(placeholder="Qty", description="Qty:")
price = widgets.FloatText(placeholder="Price", description="Price:")
output = widgets.Output()

add_button = widgets.Button(description="Add/Update", button_style='success')
view_button = widgets.Button(description="View Stock", button_style='info')
clear_button = widgets.Button(description="Clear All", button_style='danger')

# Functions
def add_item(b):
    name = item_name.value.strip()
    qty = quantity.value
    prc = price.value
    if not name:
        return
    for item in stock_data:
        if item['Item'].lower() == name.lower():
            item['Quantity'] += qty
            item['Price'] = prc  # Update price if needed
            break
    else:
        stock_data.append({'Item': name, 'Quantity': qty, 'Price': prc})
    item_name.value = ''
    quantity.value = 0
    price.value = 0.0
    display_stock()

def display_stock(b=None):
    with output:
        clear_output()
        if stock_data:
            df = pd.DataFrame(stock_data)
            display(df)
        else:
            print("Stock is empty.")

def clear_stock(b):
    stock_data.clear()
    with output:
        clear_output()
        print("Stock cleared.")

# Bind buttons
add_button.on_click(add_item)
view_button.on_click(display_stock)
clear_button.on_click(clear_stock)

# Layout
form = widgets.VBox([item_name, quantity, price, add_button, view_button, clear_button])
display(form, output)
```
